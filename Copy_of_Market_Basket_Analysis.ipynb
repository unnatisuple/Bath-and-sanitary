```python
import pandas as pd
import numpy as np

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

import matplotlib.pyplot as plt
import seaborn as sns
```

Create Sample Dataset

```python
transactions = [
    ['Milk', 'Bread', 'Butter'],
    ['Bread', 'Eggs'],
    ['Milk', 'Bread', 'Eggs', 'Butter'],
    ['Milk', 'Eggs'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread', 'Butter', 'Eggs'],
    ['Bread', 'Eggs'],
    ['Milk', 'Butter'],
    ['Milk', 'Bread', 'Eggs'],
    ['Butter', 'Eggs']
]
```

Convert Transactions into One-Hot Encoding

```python
te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(te_array, columns=te.columns_)

basket.head()
```

Generate Frequent Itemsets

```python
frequent_itemsets = apriori(
    basket,
    min_support=0.3,
    use_colnames=True
)

frequent_itemsets
```

Sort Frequent Itemsets

```python
frequent_itemsets = frequent_itemsets.sort_values(
    by='support',
    ascending=False
)

frequent_itemsets
```

Generate Association Rules

```python
rules = association_rules(
    frequent_itemsets,
    metric='confidence',
    min_threshold=0.5
)

rules.head()
```

: Important Columns

```python
rules[['antecedents',
       'consequents',
       'support',
       'confidence',
       'lift']]
```

Sort Rules by Lift

```python
rules = rules.sort_values(
    by='lift',
    ascending=False
)

rules.head(10)
```

Visualize Top Frequent Items

```python
top_items = frequent_itemsets.copy()

top_items['itemsets'] = top_items['itemsets'].astype(str)

plt.figure(figsize=(10,5))

sns.barplot(
    x='support',
    y='itemsets',
    data=top_items.head(10)
)

plt.title("Top Frequent Itemsets")
plt.show()
```

Visualize Association Rules

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=rules,
    x='support',
    y='confidence',
    size='lift',
    hue='lift'
)

plt.title("Association Rules")
plt.show()
```

Display Top Recommendations

```python
rules[['antecedents',
       'consequents',
       'confidence',
       'lift']].head(10)
```

```python
rules['antecedents'] = rules['antecedents'].apply(
    lambda x: ', '.join(list(x))
)

rules['consequents'] = rules['consequents'].apply(
    lambda x: ', '.join(list(x))
)

rules.head()
```

Business Insights

```python
top_rules = rules.sort_values(
    by='lift',
    ascending=False
).head(5)

for i,row in top_rules.iterrows():
    print(
        f"If customer buys {row['antecedents']}, "
        f"they are likely to buy {row['consequents']} "
        f"(Confidence={row['confidence']:.2f}, Lift={row['lift']:.2f})"
    )
```
